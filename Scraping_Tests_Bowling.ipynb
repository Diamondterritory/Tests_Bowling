```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import numpy as np
from tqdm import tqdm
```

```python
bowler_df = pd.DataFrame(columns = ["Name","Overs","BPO","Maidens","Runs","Wickets",
                                    "Economy","Innings","Oppn","Opposition","Venue","Start Date"])
for page_num in tqdm(range(1,220,1)):
    url = "https://stats.espncricinfo.com/ci/engine/stats/index.html?class=1;orderby=start;page=" + str(page_num) + ";team=6;template=results;type=bowling;view=innings"
    page = requests.get(url)
    bs = BeautifulSoup(page.content,"lxml")
    
    table_body = bs.find_all("tbody")
    
    
    for i, table in enumerate(table_body):
        rows = table.find_all("tr")
        for row in rows:
            cols = row.find_all("td")
            cols = [x.text.strip() for x in cols]
            bowler_df = bowler_df.append(pd.Series([cols[0],cols[1],cols[2],cols[3],cols[4],cols[5],
                                                  cols[6],cols[7],cols[8],cols[9],cols[10],cols[11]],
                                                  index = bowler_df.columns), ignore_index = True)
    
```

```python
url = "https://www.espncricinfo.com/india/content/player/caps.html?country=6;class=1"
page = requests.get(url)
bs = BeautifulSoup(page.content,"lxml")
    
links = bs.find_all("a")

results = []
for link in links:
    results.append(str(link))
```

```python
results
```

```python
names = []
for row_num in tqdm(range(0,600,2)):
    string = results[row_num]
    rindex = string.find(".html")
    string2 = string[:rindex]
    lindex = string2.rfind("/") + 1
    id_num = string2[lindex:]
    
    rindex1 = string.find("</a>")
    string3 = string[:rindex1]
    lindex1 = string3.rfind(">") + 1
    name = string3[lindex1:]
    names.append((name,id_num))

names = pd.DataFrame(names, columns=["Name","ID"])
```

```python
def bowling_style(id_num):
    url = "https://www.espncricinfo.com/india/content/player/" + str(id_num) + ".html"
    page = requests.get(url)
    bs = BeautifulSoup(page.content,"lxml")
    
    paras = str(bs.find_all('p'))
    if (paras.find("Bowling style") != -1):
        lindex = paras.find("Bowling style") + 13
        paras1 = paras[lindex:]
        lindex1 = paras1.find("<span>") + 6
        rindex1 = paras1.find("</span")
        return paras1[lindex1:rindex1]
    else:
        return "NA"
    return paras
```

```python
names["Bowling Style"] = names["ID"].apply(bowling_style)
```

```python
spin_col = ['Legbreak',
       'Right-arm offbreak', 'Slow left-arm orthodox',
       'Legbreak googly', 'Right-arm bowler',
       'Right-arm offbreak, Legbreak', 'Right-arm medium, Legbreak',
       'Right-arm fast-medium, Right-arm offbreak',
       'Right-arm medium, Right-arm offbreak',
       'Left-arm medium, Slow left-arm orthodox',
       'Right-arm fast-medium, Right-arm offbreak, Legbreak',
       'Left-arm medium-fast, Slow left-arm orthodox',
       'Right-arm offbreak, Legbreak googly',
       'Left-arm wrist-spin']
```

```python
def pace_spin(string):
    if string in spin_col:
        return "Spin"
    else:
        return "Pace"
```

```python
names["Speed"] = names["Bowling Style"].apply(pace_spin)
```
